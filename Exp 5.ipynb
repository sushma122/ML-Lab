```python
import pandas as pd
import numpy as np
df = pd.read_csv('credit.csv')
df.head(5)
```

```python
x=(df['recreation'] == "golf").sum()
unconditional_golf = x/len(df)
print("Unconditional probability of golf",unconditional_golf*100)
```

```python
df['status_new'] = np.where(df['status'] == 'single',1,0)
df.head()
```

```python
df['Risk_new'] = np.where(df['Risk'] == 'medRisk', 1, 0)
df.head(10)
```

```python
df["count"] = 1
```

```python
df = df[['status_new', 'Risk_new', 'count']]
df.head(10)
```

```python
d = pd.pivot_table(
    df,
    values = 'count',
    index = ['status_new'],
    columns = ['Risk_new'],
    aggfunc = np.size,
    fill_value = 0
)
```

```python
print(d)
```

```python
pa = (d. at[1, 0] + d.at[1, 1]) / (d.at[1, 0] + d.at[1,1] + d.at[0, 0]  + d.at[0, 1])
pb = (d. at[0, 1] + d.at[1, 1]) / (d.at[1, 0] + d.at[1,1] + d.at[0, 0]  + d.at[0, 1])
paandpb = d.at[1,1] / (d.at[1, 0] + d.at[1,1] + d.at[0, 0] + d.at[0, 1])
pabyp = paandpb/pb
print("the conditional probabilitiy of single given by medRisk",pabyp*100)
```
